```python
import brax
from brax import envs
import jax
import jax.numpy as jnp
import numpy as np
```

```python
env = envs.create(env_name="halfcheetah")
```

```python
# Reset the environment
rng = jax.random.PRNGKey(0)
state = env.reset(rng)

# Run for 100 steps
for _ in range(100):
    # Generate a random action
    action = jax.random.uniform(jax.random.PRNGKey(1), (env.action_size,), minval=-1.0, maxval=1.0)
    
    # Step through the environment
    state = env.step(state, action)
    
    # Print observations and reward
    print("Observations:", state.obs)
    print("Reward:", state.reward)
```
